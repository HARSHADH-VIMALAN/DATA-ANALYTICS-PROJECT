```python
#import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('/content/honey_purity_test11.csv' )
```

```python
# Check for missing data
print("Missing values in each column:")
print(data.isnull().sum())
```

```python
# Correlation heatmap
plt.figure(figsize=(10, 6))
# Exclude non-numeric columns from correlation calculation
correlation = data.select_dtypes(include=['number']).corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Boxplot for Purity
plt.subplot(3, 2, 1)
sns.boxplot(y='Purity (%)', data=data, color='r')
plt.title('Purity (%) - Outlier Detection')
```

```python
# Scatterplot: Purity vs Moisture Content
plt.subplot(3, 2, 2)
sns.scatterplot(x='Moisture Content (%)', y='Purity (%)', data=data, color='green')
plt.title('Purity vs Moisture Content (%)')
```

```python
# Scatterplot: Purity vs Reducing Sugar
plt.subplot(3, 2, 3)
sns.scatterplot(x='Reducing Sugar (%)', y='Purity (%)', data=data, color='orange')
plt.title('Purity vs Reducing Sugar (%)')
```

```python
# Scatterplot: Purity vs Sucrose Content
plt.subplot(3, 2, 4)
sns.scatterplot(x='Sucrose Content (%)', y='Purity (%)', data=data, color='red')
plt.title('Purity vs Sucrose Content (%)')
```

```python
# Scatterplot: Purity vs HMF
plt.subplot(3, 2, 5)
sns.scatterplot(x='HMF (mg/kg)', y='Purity (%)', data=data, color='blue')
plt.title('Purity vs HMF (mg/kg)')
```

```python
# Scatterplot: Purity vs Electrical Conductivity
plt.subplot(3, 2, 6)
sns.scatterplot(x='Electrical Conductivity (mS/cm)', y='Purity (%)', data=data, color='purple')
plt.title('Purity vs Electrical Conductivity (mS/cm)')

plt.tight_layout()
plt.show()
```

```python
#histogram
sns.histplot(data['Purity (%)'], bins=20, kde=True, color='r')
plt.title('Distribution of Purity (%)')
plt.show()
```

```python
#kdeplot
sns.kdeplot(data,x='pH',multiple='stack', color='r')
plt.title('Density Plot of Purity (%)')
plt.show()
```

```python
#ecdfplot
sns.ecdfplot(data['Purity (%)'], color='r')
plt.title('ECDF Plot of Purity (%)')
plt.show()
```

```python
#lineplot
sns.lineplot(x=data.index, y=data['Purity (%)'], color='r')
plt.title('Line Plot of Purity (%)')
plt.show()
```

```python
#regression plot
sns.regplot(x=data['Moisture Content (%)'], y=data['Purity (%)'], color='r')
plt.title('Regression Plot of Moisture Content vs Purity (%)')
plt.show()
```

```python
#pair plot
sns.pairplot(data)
plt.show()
```

```python
#barplot
sns.barplot(x='Reducing Sugar (%)', y='Purity (%)', data=data, color='r')
plt.title('Bar Plot of Reducing Sugar (%) vs Purity (%)')
plt.show()
```

```python
#swarm plot
sns.swarmplot(x='Sucrose Content (%)', y='Reducing Sugar (%)', data=data, color='r')
plt.title('Sucrose Content (%) vs Reducing Sugar (%)')
plt.show()
```

```python
#strip plot
sns.stripplot(x='Moisture Content (%)', y='Purity (%)', data=data, color='r')
plt.title('Strip Plot of Moisture Content (%) vs Purity (%)')
plt.show()
```

```python
#faceit
sns.lmplot(x='Moisture Content (%)', y='Purity (%)', data=data, hue='Reducing Sugar (%)', fit_reg=False)
plt.title('Linear Regression Plot of Moisture Content vs Purity (%)')
plt.show()
```

```python
#joint plot
sns.jointplot(x='Moisture Content (%)', y='Purity (%)', data=data, kind='reg', color='r')
plt.title('Joint Plot of Moisture Content vs Purity (%)')
plt.show()
```
